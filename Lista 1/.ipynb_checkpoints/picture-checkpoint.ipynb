```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import time
import urllib

%matplotlib inline
```

```python
def log_progress(sequence, every=None, size=None, name='Items'):
    from ipywidgets import IntProgress, HTML, VBox
    from IPython.display import display

    is_iterator = False
    if size is None:
        try:
            size = len(sequence)
        except TypeError:
            is_iterator = True
    if size is not None:
        if every is None:
            if size <= 200:
                every = 1
            else:
                every = int(size / 200)     # every 0.5%
    else:
        assert every is not None, 'sequence is iterator, set every'

    if is_iterator:
        progress = IntProgress(min=0, max=1, value=1)
        progress.bar_style = 'info'
    else:
        progress = IntProgress(min=0, max=size, value=0)
    label = HTML()
    box = VBox(children=[label, progress])
    display(box)

    index = 0
    try:
        for index, record in enumerate(sequence, 1):
            if index == 1 or index % every == 0:
                if is_iterator:
                    label.value = '{name}: {index} / ?'.format(
                        name=name,
                        index=index
                    )
                else:
                    progress.value = index
                    label.value = u'{name}: {index} / {size}'.format(
                        name=name,
                        index=index,
                        size=size
                    )
            yield record
    except:
        progress.bar_style = 'danger'
        raise
    else:
        progress.bar_style = 'success'
        progress.value = index
        label.value = "{name}: {index}".format(
            name=name,
            index=str(index or '?')
        )
```

```python
def pbil(F, I, N, LEN, p1, p2, p3):
    t0 = time.time()
    EVAL = np.zeros(N)
    bests = np.zeros(I)
    mids = np.zeros(I)
    lows = np.zeros(I)
    best_one = np.zeros((I, LEN))
    probs = np.zeros((I, LEN))
    p = np.full(LEN, 0.5)
    for iteration in log_progress(range(I)):
        P = np.random.uniform(0, 1, (N, LEN)) < p
        EVAL = np.apply_along_axis(F, 1, P)
        probs[iteration] = p.copy()
        
        # best selection
        x = P[np.argmax(EVAL), :]
        p = p * (1 - p1) + x * p1
        
        # random mutation
        mut_prob = np.random.uniform(0,1, LEN)
        mut = p * (1-p3) + (np.random.uniform(0, 1, LEN) < np.repeat(0.5, LEN)) * p3
        p -= p * (mut_prob < p2)
        p += (mut_prob < p2) * (mut)
        
        bests[iteration] = EVAL.max()
        mids[iteration] = EVAL.mean()
        lows[iteration] = EVAL.min()
        best_one[iteration] = x
        if (iteration % 10 == 0):
            print ("iteration nr:", iteration, "time:", time.time() - t0)
    np.save("picture/bests", bests)
    np.save("picture/mids", mids)
    np.save("picture/lows", lows)
    np.save("picture/best_one", best_one)
    np.save("picture/probs", probs)
    return {'best':bests, 'mid':mids, 'low':lows, 'probs':probs, 'best_one':best_one}
```

```python
agents = np.loadtxt("ClassificationRules.txt")
reference = np.loadtxt("ImageExpertReduced.txt")
```

```python
rn = np.arange(266) + 1

def poll(ind):
    res = stats.mode(agents[ind, :])
    return np.sum(res == reference)
```

## Weighted version

```python
def random_population(N, p):
    return np.clip(np.random.normal(p, 0.15, size=(N, p.shape[0])), 0, 1)
```

```python
def weighted_dest(ind):
    def single_point(votes):
        sums = np.array([0, np.sum((votes == 1.0) * ind), np.sum((votes == 2.0) * ind), np.sum((votes == 3.0) * ind)])
        return np.argmax(sums)
    return np.sum(np.apply_along_axis(single_point, 0, agents) == reference)
```

```python
def pbilWeighted(F, I, N, LEN, p1, p2, p3):
    t0 = time.time()
    EVAL = np.zeros(N)
    bests = np.zeros(I)
    mids = np.zeros(I)
    lows = np.zeros(I)
    best_one = np.zeros((I, LEN))
    probs = np.zeros((I, LEN))
    p = np.full(LEN, 0.5)
    for iteration in log_progress(range(I)):
        P = random_population(N, p)
        EVAL = np.apply_along_axis(F, 1, P)
        probs[iteration] = p.copy()
        
        # best selection
        x = P[np.argmax(EVAL), :]
        p = p * (1 - p1) + x * p1
        
        # random mutation
        mut_prob = np.random.uniform(0,1, LEN)
        mut = p * (1-p3) + (np.random.uniform(0, 1, LEN) < np.repeat(0.5, LEN)) * p3
        p -= p * (mut_prob < p2)
        p += (mut_prob < p2) * (mut)
        
        bests[iteration] = np.max(EVAL)
        mids[iteration] = np.mean(EVAL)
        lows[iteration] = np.min(EVAL)
        best_one[iteration] = x.copy()
        if (iteration % 10 == 0):
            print ("iteration nr:", iteration, "time:", time.time() - t0)
    np.save("picture/bests_weight", bests)
    np.save("picture/mids_weight", mids)
    np.save("picture/lows_weight", lows)
    np.save("picture/best_one_weight", best_one)
    np.save("picture/probs_weight", probs)
    return {'best':bests, 'mid':mids, 'low':lows, 'probs':probs, 'best_one':best_one}
```

```python
WEIGHTED_ITERATIONS = 100
res2 = pbilWeighted(weighted_dest, WEIGHTED_ITERATIONS, 50, 266, 0.2, 0.05, 0.02)
```

```python
plt.figure(figsize=(15,10))
plt.plot(res2['best'])
plt.plot(res2['mid'])
plt.plot(res2['low'])
plt.axis([0, WEIGHTED_ITERATIONS, 8000, 9350])
plt.grid(True)
plt.show()
```
